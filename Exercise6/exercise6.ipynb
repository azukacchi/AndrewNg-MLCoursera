Gaussian Kernel

```python
def gaussianKernel(x1, x2, sigma):
    atas = np.sum((x1 - x2)**2)
    bawah = 2*(sigma**2)
    return np.exp(-atas/bawah)
```

```python
x1 = np.array([1.0, 2.0, 1.0])
x2 = np.array([0.0, 4.0, -1.0])
sigma = 2

gaussianKernel(x1, x2, sigma)
```

```python
mat3 = loadmat('ex6data3.mat')
X3 = mat3['X']
y3 = mat3['y']
Xval = mat3['Xval']
yval = mat3['yval']
print(X3.shape)
print(y3.shape)
print(Xval.shape)
print(yval.shape)
```

```python
pos = y3==1
neg = y3==0
plt.scatter(X3[pos[:,0],0], X3[pos[:,0],1], marker='+')
plt.scatter(X3[neg[:,0],0], X3[neg[:,0],1], marker='o')
plt.show()
```

```python
def dataset3Params(X, y, Xval, yval, vals):
    best_score = 0
    best_params = {'C':None, 'gamma':None}

    for i in vals:
        C = i
        for j in vals:
            gamma = 1/j
            svc_cls = SVC(C=C, gamma=gamma)
            svc_cls.fit(X,y)
            prediction = svc_cls.predict(Xval)
            score = svc_cls.score(Xval, yval)
            if score > best_score:
                best_score = score
                best_params['C'] = C
                best_params['gamma'] = gamma
    return best_params['C'], best_params['gamma']
```

```python
vals = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
best_C, best_gamma = dataset3Params(X3, y3, Xval, yval, vals)
svc_cls4 = SVC(C=best_C, gamma=best_gamma)
svc_cls4.fit(X3, y3)
svc_cls4.score(X3, y3)
```

```python
pos = y3==1
neg = y3==0
plt.scatter(X3[pos[:,0],0], X3[pos[:,0],1], marker='+')
plt.scatter(X3[neg[:,0],0], X3[neg[:,0],1], marker='o')

# plotting the decision boundary
X_1,X_2 = np.meshgrid(np.linspace(X3[:,0].min(),X3[:,1].max(),num=100),np.linspace(X3[:,1].min(),X3[:,1].max(),num=100))

plt.contour(X_1,X_2, svc_cls4.predict(np.array([X_1.ravel(),X_2.ravel()]).T).reshape(X_1.shape),1,colors='b')
plt.xlim(-0.6,0.3)
plt.ylim(-0.7,0.5)

plt.show()
```

## Spam Classification

```python
import re
from nltk.stem import PorterStemmer

sample = open("emailSample1.txt","r").read()
```

```python
vocab = open("vocab.txt","r").read().split('\n')[:-1]

vocab_dict = {}
for word in vocab:
    val, key = word.split('\t')[:]
    vocab_dict[key] = val

print(len(vocab_dict))
list(vocab_dict.keys())[:5]
```

```python
sample
```

```python
def processEmail(sample,vocab_dict):
    """
    Preprocesses the body of an email and returns a list of indices of the words contained in the email. 
    """
    # Lower case
    sample = sample.lower()
    
    # Handle numbers
    sample = re.sub('[0-9]+','number',sample)
    
    # Handle URLS
    pattern = r'@^(http\:\/\/|https\:\/\/)?([a-z0-9][a-z0-9\-]*\.)+[a-z0-9][a-z0-9\-]*$@i'
    sample = re.sub(pattern, 'httpaddr', sample)
        
    # Handle Email Addresses
    pattern = '[^\s]+@[^\s]+'
    sample = re.sub(pattern, 'emailaddr', sample)
        
    # Handle $ sign
    sample = re.sub('[$]+', 'dollar', sample)
        
    # Strip all special characters
    pattern = r'[@><$-/:-?{-~!"^_`\[\]\r]+'
    sample = re.sub(pattern, '', sample)
    sample = re.sub('\n', ' ', sample)    
    
    # Stem the word
    stemmer = PorterStemmer()
    sample = [stemmer.stem(token) for token in sample.split(' ')]
    sample = ' '.join(sample)
    
    # Process the email and return word_indices
    
    word_indices=[]
    
    for word in sample.split():
        if len(word) > 1 and word in vocab_dict:
            word_indices.append(int(vocab_dict[word]))
    
    return word_indices
```

```python
word_indices = processEmail(sample, vocab_dict)
word_indices
```

```python
def emailFeatures(word_indices, vocab_dict):
    k = len(vocab_dict)
    features = np.zeros((k,1))
    for i in word_indices:
        features[i-1] = 1
    return features
```

```python
features = emailFeatures(word_indices, vocab_dict)
print('Length of feature vector: ',len(features))
print('Number of non-zero entries: ',np.sum(features))
```
